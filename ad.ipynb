```python
import pandas as pd
import numpy as np

# Original DataFrame
data = pd.DataFrame({
    'Brand': ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Maruti', 'Hyundai',
              'Renault', 'Tata', 'Maruti'],
    'Year': [2012, 2014, 2011, 2015, 2012, 2016, 2014, 2018, 2019],
    'Kms Driven': [50000, 30000, 60000, 25000, 10000, 46000, 31000, 15000, 12000],
    'City': ['Gurgaon', np.nan, 'Mumbai', 'Delhi', 'Mumbai', 'Delhi',
             'Mumbai', 'Chennai', 'Ghaziabad'],
    'Mileage': [28, 27, 25, 26, 28, 29, 24, 21, 24]
})

# 1. Display the DataFrame
display(data)

# 2. Filter rows where Brand is 'Maruti' and Year is 2012
maruti_2012 = data.loc[(data.Brand == 'Maruti') & (data.Year == 2012)]
display(maruti_2012)

# 3. Filter rows where Brand is 'Hyundai' and Kms Driven > 30000
hyundai_above_30k = data[(data.Brand == 'Hyundai') & (data['Kms Driven'] > 30000)]
display(hyundai_above_30k)

# 4. Filter rows where City is 'Delhi' and Kms Driven > 30000
delhi_above_30k = data[(data.City == 'Delhi') & (data['Kms Driven'] > 30000)]
display(delhi_above_30k)

# 5. Creating another DataFrame for join operations
other_data = pd.DataFrame({
    'Brand': ['Maruti', 'Hyundai', 'Tata', 'Mahindra', 'Renault'],
    'Customer Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Discount Amount': [5000, 3000, 2000, 1500, 1000]
})

# Display the second DataFrame
display(other_data)

# 6. Inner Join on 'Brand'
inner_join = pd.merge(data, other_data, on='Brand', how='inner')
display(inner_join)

# 7. Outer Join on 'Brand'
outer_join = pd.merge(data, other_data, on='Brand', how='outer')
display(outer_join)

# 8. Left Join on 'Brand'
left_join = pd.merge(data, other_data, on='Brand', how='left')
display(left_join)

# 9. Right Join on 'Brand'
right_join = pd.merge(data, other_data, on='Brand', how='right')
display(right_join)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Data with missing values
data = {
    'Age': [25, None, 30, 35, None, 4.5],
    'Height': [5.5, 6.1, 5.9, None, 5.7, 6.2],
    'HelmetStatus': [1, None, 1, 0, 1, 1],
    'Gender': ['M', 'F', 'M', 'None', 'F', 'M'],
    'Experience': [2, 4, 6, 1, None, None]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame with Missing values:")
print(df)

# Creating copies of the original DataFrame
df_filled_mean = df.copy()
df_filled_median = df.copy()
df_filled_mode = df.copy()

# Filling missing values with mean for 'Age', 'Height', and 'Experience'
df_filled_mean['Age'] = df_filled_mean['Age'].fillna(df['Age'].mean())
df_filled_mean['Experience'] = df_filled_mean['Experience'].fillna(df['Experience'].mean())

# Convert the columns to integers after filling with mean
df_filled_mean['Age'] = df_filled_mean['Age'].astype(int)
df_filled_mean['Experience'] = df_filled_mean['Experience'].astype(int)

# Filling missing values with median for 'Age', 'Height', and 'Experience'
df_filled_median['Age'] = df_filled_median['Age'].fillna(df['Age'].median())
df_filled_median['Height'] = df_filled_median['Height'].fillna(df['Height'].median())
df_filled_median['Experience'] = df_filled_median['Experience'].fillna(df['Experience'].median())

# Convert the columns to integers after filling with median
df_filled_median['Age'] = df_filled_median['Age'].astype(int)
df_filled_median['Experience'] = df_filled_median['Experience'].astype(int)

# Filling missing values with mode for 'HelmetStatus' and 'Gender'
df_filled_mode['HelmetStatus'] = df_filled_mode['HelmetStatus'].fillna(df['HelmetStatus'].mode()[0])
df_filled_mode['Gender'] = df_filled_mode['Gender'].fillna(df['Gender'].mode()[0])

# Display the DataFrames after imputation
print("\nDataFrame after filling with Mean:")
print(df_filled_mean)

print("\nDataFrame after filling with Median:")
print(df_filled_median)

print("\nDataFrame after filling with Mode:")
print(df_filled_mode)
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Assuming 'cars' is your DataFrame containing the car data
# Read the data if you haven't already
cars = pd.read_csv("Cars.csv")  # Replace 'Cars.csv' with the actual file name if needed

# Define and fit the OLS model
# Adjust the formula if needed to include the correct predictor variables
model = smf.ols('MPG ~ HP + WT + VOL + SP', data=cars).fit()

# Now you can create the regression plots
fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, "WT", fig=fig)
plt.show()

fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, "SP", fig=fig) # Changed 'sp' to 'SP' to match column name
plt.show()

fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, "VOL", fig=fig)
plt.show()

fig = plt.figure(figsize=(15, 8))
fig = sm.graphics.plot_regress_exog(model, "HP", fig=fig)
plt.show()
```

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm
import numpy as np

import statsmodels.formula.api as smf

model = smf.ols('MPG ~ HP + WT + VOL + SP', data=cars).fit()


model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance

fig = plt.figure(figsize=(20, 7))
plt.stem(np.arange(len(cars)), np.round(c, 3))
plt.xlabel("Row Index")
plt.ylabel("Cooks Distance")
plt.show()
```
